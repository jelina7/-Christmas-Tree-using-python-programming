```python
height = 7

# Star
print(" " * (height - 1) + "‚≠ê")

# Tree
for i in range(height):
    print(" " * (height - i - 1) + "*" * (2 * i + 1))

# Decorations
for _ in range(2):
    print(" " * (height - 2) + "o o")

# Trunk
for _ in range(3):
    print(" " * (height - 1) + "|")

# Base
print(" " * (height - 2) + "===")

# Christmas Wish
print("üéÑ‚ú® Merry Christmas! ‚ú®üéÑ")
print("May your life be filled with joy, peace, and blessings ‚ù§Ô∏è")
```
